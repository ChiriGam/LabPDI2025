# Mejorar calidad de imagen a color
Importación de librerías:


*   `cv2` es la librería OpenCV, utilizada para procesamiento de imágenes.
*   `matplotlib.pyplot` se usa para mostrar imágenes y gráficos.






```python
import cv2
import matplotlib.pyplot as plt
```

Carga de imagen:

* `cv2.imread()` lee la imagen desde la ruta indicada y la carga en formato BGR (formato por defecto en OpenCV).

* `cv2.cvtColor()` convierte la imagen de BGR a RGB para que la visualización con Matplotlib sea correcta, ya que Matplotlib usa RGB.



```python
# Cargar imagen oscura
img_bgr = cv2.imread('/content/Im004_ACRIMA.jpg')
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
```

Conversión de espacio de color:

* La imagen se convierte de BGR a HSV (Hue, Saturation, Value).

* Esto permite trabajar con el canal de brillo (Value) separadamente.

```python
# Convertir a HSV
img_hsv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)
```

Separación de canales:

Se extraen los tres canales individuales:

* `h`: tono (color)

* `s`: saturación (intensidad del color)

* `v`: valor (brillo)

```python
# Separar canales H, S y V
h, s, v = cv2.split(img_hsv)
```

Ecualización del histograma:

* Se aplica la ecualización solo al canal de brillo v.

* Esta técnica mejora el contraste y la luminosidad, haciendo visibles detalles en zonas oscuras.

```python
# Ecualizar solo el canal de valor (brillo)
v_eq = cv2.equalizeHist(v)
```

Reconstrucción de la imagen HSV:

* Se combinan los canales `h` y `s` originales con el canal `v_eq` ecualizado para formar una nueva imagen HSV mejorada.

```python
# Combinar canales con el V ecualizado
img_hsv_eq = cv2.merge([h, s, v_eq])
```

Conversión para visualización:

* Primero se convierte la imagen mejorada de HSV a BGR para que OpenCV pueda manejarla correctamente.

* Luego, se convierte de BGR a RGB para mostrarla con Matplotlib.

```python
# Convertir de nuevo a BGR y luego a RGB para visualización
img_eq_bgr = cv2.cvtColor(img_hsv_eq, cv2.COLOR_HSV2BGR)
img_eq_rgb = cv2.cvtColor(img_eq_bgr, cv2.COLOR_BGR2RGB)
```

Visualización:

Se crea una figura con dos subgráficos:

* El primero muestra la imagen original.

* El segundo muestra la imagen mejorada tras la ecualización del canal V.

Se desactivan los ejes para una mejor presentación visual.

* `plt.tight_layout()` ajusta el espaciado entre gráficos para que no se solapen.

* `plt.show()` muestra la ventana con las imágenes.

```python
# Mostrar imágenes original y mejorada
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title('Imagen original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img_eq_rgb)
plt.title('Imagen mejorada (Ecualización V)')
plt.axis('off')

plt.tight_layout()
plt.show()
```

Se repite el mismo procedimiento para las otras dos imagenes.

```python
# Cargar imagen oscura
img2_bgr = cv2.imread('/content/person78_virus_140.jpeg')
img2_rgb = cv2.cvtColor(img2_bgr, cv2.COLOR_BGR2RGB)

# Convertir a HSV
img2_hsv = cv2.cvtColor(img2_bgr, cv2.COLOR_BGR2HSV)

# Separar canales H, S y V
h2, s2, v2 = cv2.split(img2_hsv)

# Ecualizar solo el canal de valor (brillo)
v2_eq = cv2.equalizeHist(v2)

# Combinar canales con el V ecualizado
img2_hsv_eq = cv2.merge([h2, s2, v2_eq])

# Convertir de nuevo a BGR y luego a RGB para visualización
img2_eq_bgr = cv2.cvtColor(img2_hsv_eq, cv2.COLOR_HSV2BGR)
img2_eq_rgb = cv2.cvtColor(img2_eq_bgr, cv2.COLOR_BGR2RGB)
```

```python
# Mostrar imágenes original y mejorada
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img2_rgb)
plt.title('Imagen original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img2_eq_rgb)
plt.title('Imagen mejorada (Ecualización V)')
plt.axis('off')

plt.tight_layout()
plt.show()
```
