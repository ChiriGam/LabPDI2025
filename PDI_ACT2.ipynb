# Parte I.- Procesamiento geométrico
## *Ejercicio 1: Traslación*


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#Cargar Imagen
img = cv2.imread('Resonancia1.jpg', cv2.IMREAD_GRAYSCALE)

#Obtener Dimensiones de la Imagen
alto, ancho = img.shape

#Definir Matriz de Traslación
tx, ty = 50, 30  # Desplazamiento en X e Y
M = np.float32([[1, 0, tx], [0, 1, ty]])
txf, tyf = 20.5, 15.5  # Desplazamiento en X e Y con decimales
Mf = np.float32([[1, 0, txf], [0, 1, tyf]])

#Aplicar Transformación
img_t = cv2.warpAffine(img, M, (ancho, alto))
img_tf = cv2.warpAffine(img, Mf, (ancho, alto), flags=cv2.INTER_LINEAR)

#Visualizar Imágenes
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.imshow(img, cmap="gray")
plt.title("Imagen Original")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(img_t, cmap="gray")
plt.title("Imagen Trasladada")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(img_tf, cmap="gray")
plt.title("Imagen Trasladada (Decimales)")
plt.axis("off")

plt.show()
```

## *Ejercicio 2: Rotación*

```python
# Definir rotación (30° alrededor del centro de la imagen)
angulo = 45
centro = (ancho // 2, alto // 2)  # Centro de la imagen
escala = 1.0  # Sin cambios en el tamaño

M_rotacion = cv2.getRotationMatrix2D(centro, angulo, escala)

# Aplicar rotación sobre la imagen trasladada
img_r = cv2.warpAffine(img, M_rotacion, (ancho, alto))

#Visualizar Imágenes
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img, cmap="gray")
plt.title("Imagen Original")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(img_r, cmap="gray")
plt.title("Imagen Rotada")
plt.axis("off")

plt.show()
```

## *Ejercicio 3: Escala*

```python
# Definir escala
angulo = 0
centro = (ancho // 2, alto // 2)  # Centro de la imagen
escala1 = 1.5
escala2 = 0.5

M_escala1 = cv2.getRotationMatrix2D(centro, angulo, escala1)
M_escala2 = cv2.getRotationMatrix2D(centro, angulo, escala2)

# Aplicar rotación sobre la imagen trasladada
img_e1 = cv2.warpAffine(img, M_escala1, (ancho, alto))
img_e2 = cv2.warpAffine(img, M_escala2, (ancho, alto))

#Visualizar Imágenes
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.imshow(img, cmap="gray")
plt.title("Imagen Original")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(img_e1, cmap="gray")
plt.title("Imagen Escalada al 150%")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(img_e2, cmap="gray")
plt.title("Imagen Escalada al 50%")
plt.axis("off")

plt.show()
```

# Parte II.- Ecualización de histograma
## *Ejercicio 1: Ecualización*


```python
#Cargar Imagen
img2 = cv2.imread('Resonancia2.jpg', cv2.IMREAD_GRAYSCALE)

# Aplicar ecualización de histograma
img_ez = cv2.equalizeHist(img2)

# Calcular histogramas antes y después
hist_img2 = cv2.calcHist([img2], [0], None, [256], [0,256])
hist_ez = cv2.calcHist([img_ez], [0], None, [256], [0,256])

# Mostrar imágenes e histogramas
plt.figure(figsize=(10,5))

plt.subplot(2,2,1)
plt.imshow(img2, cmap="gray")
plt.title("Imagen Original")
plt.axis("off")

plt.subplot(2,2,2)
plt.plot(hist_img2, color="black")
plt.title("Histograma Original")
plt.xlabel("Intensidad")
plt.ylabel("Frecuencia")

plt.subplot(2,2,3)
plt.imshow(img_ez, cmap="gray")
plt.title("Imagen Ecualizada")
plt.axis("off")

plt.subplot(2,2,4)
plt.plot(hist_ez, color="black")
plt.title("Histograma Ecualizado")
plt.xlabel("Intensidad")
plt.ylabel("Frecuencia")

plt.tight_layout()
plt.show()
```
